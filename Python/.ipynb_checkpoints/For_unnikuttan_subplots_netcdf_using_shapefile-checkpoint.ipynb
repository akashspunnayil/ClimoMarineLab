# FUNCTION TO PLOT

```python
import xarray as xr
import geopandas as gpd
from shapely.geometry import mapping

def clip_precipitation_data(nc_file_path, shapefile_path):
    """
    Clip precipitation data from a NetCDF file using a shapefile.

    Parameters:
        nc_file_path (str): Path to the NetCDF file containing precipitation data.
        shapefile_path (str): Path to the shapefile used for clipping.

    Returns:
        xr.Dataset: Clipped precipitation data as an xarray Dataset.
    """
    # Load the precipitation data
    dataset = xr.open_dataset(nc_file_path)

    # Load the shapefile using geopandas
    shapefile = gpd.read_file(shapefile_path)

    # Set spatial dimensions and CRS
    dataset.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace=True)
    dataset.rio.write_crs("epsg:4326", inplace=True)  # Assuming WGS84

    # Clip the precipitation data using the shapefile
    clipped_data = dataset.rio.clip(
        shapefile.geometry.apply(mapping), shapefile.crs, drop=False
    )
#     clipped = precip_data.rio.clip(india_shapefile.geometry.apply(mapping), india_shapefile.crs, \
#                                drop=False)

    return clipped_data
```

```python
# File paths
nc_file_path = './data/precip.mon.ltm.1991-2020-INDIA_025deg.nc'
shapefile_path = './data/India Shapefile With Kashmir/India Shape/india_st.shp'

# Call the function
clipped_data = clip_precipitation_data(nc_file_path, shapefile_path)

# Assuming clipped_data, india_gdf, and other required variables are defined earlier

# Subset data for June, July, August, and September
clipped_june_data = clipped_data['PRECIPITATION'][5, :, :] / 10
clipped_july_data = clipped_data['PRECIPITATION'][6, :, :] / 10
clipped_august_data = clipped_data['PRECIPITATION'][7, :, :] / 10
clipped_september_data = clipped_data['PRECIPITATION'][8, :, :] / 10

# Now you can proceed with plotting or any other operations using the clipped data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from matplotlib.ticker import FuncFormatter
import math

def plot_contour_subplots(data_list, common_title, titles, xticks=None, yticks=None, xlim=None, ylim=None, levels=50, cmap='plasma', \
                          figsize=(12, 12), subplot_kw={'projection': ccrs.PlateCarree()}, saveplot=True, frmt='.png'):
    
    """
    Plot contour subplots with given data and titles.
    
    Parameters:
        data_list (list of tuples): List of tuples where each tuple contains (lon, lat, data).
        titles (list of str): List of titles for each subplot.
        xticks (list of arrays, optional): List of xticks for each subplot.
        yticks (list of arrays, optional): List of yticks for each subplot.
        xlim (list of tuples, optional): List of x-axis limits for each subplot.
        ylim (list of tuples, optional): List of y-axis limits for each subplot.
        levels (int, optional): Number of contour levels.
        cmap (str, optional): Colormap for the contours.
        figsize (tuple, optional): Figure size.
        subplot_kw (dict, optional): Keyword arguments for subplots.
        saveplot (bool, optional): Whether to save the plot.
        frmt (str, optional): Format for saving the plot.
    """
    
    def format_lon_tick(x, pos):
        return f"{x}°E"
    
    def format_lat_tick(y, pos):
        return f"{y}°N"
    
    plt.rcParams.update({'font.size': 18})
    
    n_subplots = len(data_list)
    nrows = n_subplots // 2 + n_subplots % 2
    ncols = 2 if n_subplots > 1 else 1
    
    fig_width = ncols * 6  # Assuming 6 inches width for each subplot
    fig_height = nrows * 6  # Assuming 6 inches height for each subplot
    fig, axs = plt.subplots(max(nrows, 1), max(ncols, 1), figsize=(fig_width, fig_height), subplot_kw=subplot_kw)

    if n_subplots == 1:
        axs = np.array([axs])  # Convert to array for consistent indexing
    else:
        axs = axs.flatten()  # Flatten in case of multiple subplots   
        
#    # Remove the last subplot if it's odd and greater than 1
    if n_subplots > 1 and n_subplots % 2 != 0:
            axs[-1].remove()

    for idx, ((lon, lat, data), title) in enumerate(zip(data_list, titles)):
        ax = axs[idx]
        contour = ax.contourf(lon, lat, data, levels=levels, cmap=cmap, extend="both")
        ax.set_title(title, fontsize=18)
        
        # Set xlim for bottom subplots and ylim for left subplots
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
        if idx // ncols == nrows - 1:  # bottom subplots
            if xlim is not None:
                ax.set_xlim(xlim)
            if xticks is not None:
                ax.set_xticks(xticks)
        if idx % ncols == 0:  # left subplots
            if ylim is not None:
                ax.set_ylim(ylim)
            if yticks is not None:
                ax.set_yticks(yticks)
        
        india_gdf.boundary.plot(ax=ax, color=None, edgecolor='black', linewidth=1)
        ax.xaxis.set_major_formatter(FuncFormatter(format_lon_tick))
        ax.yaxis.set_major_formatter(FuncFormatter(format_lat_tick))
        
        # Set labels
        ax.set_xlabel('Longitude', fontsize=18)
        ax.set_ylabel('Latitude', fontsize=18)
        
    # Add common title
    plt.suptitle(common_title, fontsize=22, y=0.95)
    
    # Adjust layout
    plt.tight_layout()
    fig.subplots_adjust(hspace=0.2, wspace=0.01, bottom=0.15, top=0.85)  

    # Create a common colorbar
    cbar_ax = fig.add_axes([0.1, 0.008, 0.9, 0.03])  # [left, bottom, width, height]
    cbar = fig.colorbar(contour, cax=cbar_ax, orientation='horizontal', pad=0.9, aspect=90,\
                       ticks=np.linspace(np.min(contour.levels), np.max(contour.levels), 5))
    cbar.set_label('Precipitation (cm/day)')
    
    if saveplot:
        fig.savefig(f'./plot.{frmt}', bbox_inches='tight', dpi=300)
        print('plot saved to current directory')
        
    else:
        print('plot not saved')
    
    plt.show()

```
