```python
import os
import glob
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Define the directory containing the NetCDF files
directory = './data/'

# Get the list of all NetCDF files in the directory matching the pattern
netcdf_files = glob.glob(os.path.join(directory, 'sst_*.nc'))

# Number of files
n_files = len(netcdf_files)
n_cols = 2  # Number of columns
n_rows = (n_files + 1) // n_cols  # Calculate the number of rows needed

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, n_rows * 3))

# Flatten axes for easy iteration
axes = axes.flatten()

# Customization options
title_fontsize = 14
label_fontsize = 12
font_family = 'serif'
axis_limits = {'x': [50, 80], 'y': [8, 30]}  # Example limits, adjust as needed
# colorbar_limits = [14, 28]  # Example limits, adjust as needed

# Normalize the color scale
# norm = mcolors.Normalize(vmin=colorbar_limits[0], vmax=colorbar_limits[1])

# Loop over each NetCDF file
for i, file in enumerate(netcdf_files):
    # Load the data
    data = xr.open_dataset(file)
    
    # Select the SST data (or any other variable you are interested in)
    sst = data['SST']  # Adjust 'SST' to the variable name in your dataset
    
    # Plot the data
#     levels = np.linspace(20, 28, 10)
    contour = axes[i].contourf(data['LON2504_2959'], data['LAT1025_1366'], sst, levels=10, cmap='afmhot_r')
    
    # Add colorbar
    cbar = fig.colorbar(contour, ax=axes[i])
    cbar.set_label('SST (C)', fontsize=label_fontsize, fontfamily=font_family)
#     cbar.set_ticks(range(colorbar_limits[0], colorbar_limits[1]+1, 5))
    
    # Set axis labels
    axes[i].set_xlabel('Longitude', fontsize=label_fontsize, fontfamily=font_family)
    axes[i].set_ylabel('Latitude', fontsize=label_fontsize, fontfamily=font_family)
    
    # Set axis limits
    axes[i].set_xlim(axis_limits['x'])
    axes[i].set_ylim(axis_limits['y'])
    
    # Set title
    axes[i].set_title(f'SST Data from File {i+1}', fontsize=title_fontsize, fontfamily=font_family)

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.subplots_adjust(hspace=0.4, wspace=0.3)  # Adjust spacing as needed

plt.savefig('./subplots.png', bbox_inches='tight', dpi=300)
# Adjust layout
# plt.tight_layout()
plt.show()
```
