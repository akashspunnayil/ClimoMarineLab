## Load packages 

```python
import xarray as xr
import numpy as np
import geopandas as gpd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
```

## Load precipitation data¶

```python
# Load the entire dataset
dataPath = './data/' # change if necessary
dataset = xr.open_dataset(dataPath + 'precip.mon.ltm.1991-2020-INDIA.nc')

# Select the 'PRECIPITATION' variable
precip_var = dataset['PRECIPITATION']

print(dataset.info)
print(' ')
```

## Load shapefile¶

```python
# Load India shapefile
india_shapefile = "/media/akashs/FA22E72622E6E69B/DATA1_disk/MyGithub/Tutorials/ClimoMarineLab/Python/data/India Shapefile With Kashmir/India Shape/india_st.shp"
india_gdf = gpd.read_file(india_shapefile)
print('india_gdf')
# print(india_gdf)
india_gdf
```

## overlay land boundaries using cartopy

```python
%%time

# Subset data for June, July, August, and September
june_data = precip_var.isel(TIME=5).where((precip_var.isel(TIME=5) > 0) & (~np.isnan(precip_var.isel(TIME=5))), drop=True)
july_data = precip_var.isel(TIME=6).where((precip_var.isel(TIME=6) > 0) & (~np.isnan(precip_var.isel(TIME=6))), drop=True)
august_data = precip_var.isel(TIME=7).where((precip_var.isel(TIME=7) > 0) & (~np.isnan(precip_var.isel(TIME=7))), drop=True)
september_data = precip_var.isel(TIME=8).where((precip_var.isel(TIME=8) > 0) & (~np.isnan(precip_var.isel(TIME=8))), drop=True)

# Extract LAT39_52itude and LON27_40gitude limits
LON27_40_min = precip_var['LON27_40'].min()
LON27_40_max = precip_var['LON27_40'].max()
LAT39_52_min = precip_var['LAT39_52'].min()
LAT39_52_max = precip_var['LAT39_52'].max()

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# Define common contour levels for all subplots
levels = np.linspace(0, precip_var.max(), 11)

# Plot contour for June
cf = axs[0, 0].contourf(june_data['LON27_40'], june_data['LAT39_52'], june_data, levels=levels, cmap='viridis')#, mask=land_mask)
axs[0, 0].set_title('June')

# Plot contour for July
cf = axs[0, 1].contourf(july_data['LON27_40'], july_data['LAT39_52'], july_data, levels=levels, cmap='viridis')#, mask=land_mask)
axs[0, 1].set_title('July')

# Plot contour for August
cf = axs[1, 0].contourf(august_data['LON27_40'], august_data['LAT39_52'], august_data, levels=levels, cmap='viridis')#, mask=land_mask)
axs[1, 0].set_title('August')

# Plot contour for September
cf = axs[1, 1].contourf(september_data['LON27_40'], september_data['LAT39_52'], september_data, levels=levels, cmap='viridis')#, mask=land_mask)
axs[1, 1].set_title('September')

# Add coastlines, borders, and state boundaries
for ax in axs.flat:
#     ax.add_feature(cfeature.LAND, zorder=100, edgecolor='k', facecolor='white')  # Land with black borders
    ax.add_feature(cfeature.OCEAN, linestyle='-', zorder=100, edgecolor='k', facecolor='white')
    ax.add_feature(cfeature.BORDERS, linestyle='-', edgecolor='black', linewidth=1.0)
    ax.add_feature(cfeature.COASTLINE, linestyle='-', edgecolor='black', linewidth=1.0)
    
    # Create a feature for States/Admin 1 regions at 1:50m from Natural Earth
    states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')
    ax.add_feature(states_provinces, edgecolor='k')

    # Set labels
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

# Add colorbar with custom position
cbar = plt.colorbar(cf, ax=axs, orientation='horizontal', fraction=0.02, pad=0.1, aspect=30, location='bottom')
cbar.set_label('Precipitation (mm/day)')
cbar.ax.set_position([0.2, -0.05, 0.6, 0.02])  # Adjust the position [left, bottom, width, height]


# Add x-axis and y-axis ticks and labels
for ax in axs.flat:
    ax.set_xticks(np.arange(65, 100, 10))  # Adjust the x-axis ticks as needed
    ax.set_yticks(np.arange(0, 50, 10))   # Adjust the y-axis ticks as needed
    ax.set_xticklabels(ax.get_xticks(), fontsize=8)
    ax.set_yticklabels(ax.get_yticks(), fontsize=8)
    # Set xlim and ylim
    ax.set_xlim(LON27_40_min, LON27_40_max)
    ax.set_ylim(LAT39_52_min, LAT39_52_max)

plt.tight_layout()
plt.show()
```

## overlay land boundaries using shapefile

```python
%%time

# Assuming precip_var is an xarray Dataset

# Subset data for June, July, August, and September
june_data = precip_var.isel(TIME=5).where((precip_var.isel(TIME=5) > 0) & (~np.isnan(precip_var.isel(TIME=5))), drop=True)
july_data = precip_var.isel(TIME=6).where((precip_var.isel(TIME=6) > 0) & (~np.isnan(precip_var.isel(TIME=6))), drop=True)
august_data = precip_var.isel(TIME=7).where((precip_var.isel(TIME=7) > 0) & (~np.isnan(precip_var.isel(TIME=7))), drop=True)
september_data = precip_var.isel(TIME=8).where((precip_var.isel(TIME=8) > 0) & (~np.isnan(precip_var.isel(TIME=8))), drop=True)

# Extract latitude and longitude limits
lon_min = precip_var['LON27_40'].min()
lon_max = precip_var['LON27_40'].max()
lat_min = precip_var['LAT39_52'].min()
lat_max = precip_var['LAT39_52'].max()

# Load India shapefile
india_shapefile = "/media/akashs/FA22E72622E6E69B/DATA1_disk/MyGithub/Tutorials/ClimoMarineLab/Python/data/India Shapefile With Kashmir/India Shape/india_st.shp"
# india_shapefile = "/media/akashs/FA22E72622E6E69B/DATA1_disk/MyGithub/Tutorials/ClimoMarineLab/Python/data/India Shapefile With Kashmir/India Shape/India_Country_Boundary.shp"
india_gdf = gpd.read_file(india_shapefile)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# Define common contour levels for all subplots
levels = np.linspace(0, precip_var.max(), 11)

# Plot contour for June
cf = axs[0, 0].contourf(june_data['LON27_40'], june_data['LAT39_52'], june_data, levels=levels, cmap='viridis')
axs[0, 0].set_title('June')

# Plot contour for July
cf = axs[0, 1].contourf(july_data['LON27_40'], july_data['LAT39_52'], july_data, levels=levels, cmap='viridis')
axs[0, 1].set_title('July')

# Plot contour for August
cf = axs[1, 0].contourf(august_data['LON27_40'], august_data['LAT39_52'], august_data, levels=levels, cmap='viridis')
axs[1, 0].set_title('August')

# Plot contour for September
cf = axs[1, 1].contourf(september_data['LON27_40'], september_data['LAT39_52'], september_data, levels=levels, cmap='viridis')
axs[1, 1].set_title('September')

# Add coastlines, borders, and state boundaries
for ax in axs.flat:
    ax.add_feature(cfeature.OCEAN, linestyle='-', zorder=100, edgecolor='k', facecolor='white')
#     ax.add_feature(cfeature.BORDERS, linestyle='-', edgecolor='black', linewidth=1.0)
    ax.add_feature(cfeature.COASTLINE, linestyle='-', edgecolor='black', linewidth=1.0)
    
    # Overlay India state boundaries using the GeoDataFrame
    india_gdf.boundary.plot(ax=ax, color=None, edgecolor='white', linewidth=2)

    # Set labels
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

# Add colorbar with custom position
cbar = plt.colorbar(cf, ax=axs, orientation='horizontal', fraction=0.02, pad=0.1, aspect=30, location='bottom')
cbar.set_label('Precipitation (mm/day)')
cbar.ax.set_position([0.2, -0.05, 0.6, 0.02])  # Adjust the position [left, bottom, width, height]

# Add x-axis and y-axis ticks and labels
for ax in axs.flat:
    ax.set_xticks(np.arange(65, 100, 10))  # Adjust the x-axis ticks as needed
    ax.set_yticks(np.arange(0, 50, 10))   # Adjust the y-axis ticks as needed
    ax.set_xticklabels(ax.get_xticks(), fontsize=8)
    ax.set_yticklabels(ax.get_yticks(), fontsize=8)
    # Set xlim and ylim
    ax.set_xlim(lon_min, lon_max)
    ax.set_ylim(lat_min, lat_max)

plt.tight_layout()
plt.show()
```
