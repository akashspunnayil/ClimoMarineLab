## Load packages 

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

## Import data

```python
data_path = './' #current directory
data = xr.open_dataset(data_path+'NOAA_ERSST_V5_1981-2010.nc', decode_times=False)
print(data.info)
```

## Read variables 

```python
sst = data['sst']
latitude = data['lat']
longitude = data['lon']
time = data['time']

print(sst.shape)
print(sst.dims)
```

## A quick  plot

```python
#ploting the first time 
sst[0,:,:].plot()
```

## More customised plot 

```python
# Define a new variable for sst at first timestamp
sst1 = sst[0,:,:]

# Plot
fig = plt.figure()
plt.contourf(longitude, latitude, sst1)
plt.colorbar()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('NOAA ERSST V5 - January Climatology')
plt.show()
```

## Time series plot 

```python
# Define a new variable for sst at first timestamp
sst1 = sst[:,52,100]

# Plot
fig = plt.figure()
plt.plot(sst1)
plt.xlabel('Time')
plt.ylabel('SST')
plt.title('NOAA ERSST V5 - Monthly Climatology')
plt.show()
```

```python
# Define a new variable for sst at first timestamp
sst1 = sst[0,:,:]
import matplotlib as mpl
mpl.rcParams['font.size'] = 20  # Change 12 to your desired font size
mpl.rcParams['font.weight'] = 'bold'

# Plot
fig = plt.figure(figsize=(20, 10))
contour = plt.contourf(longitude, latitude, sst1)
cbar = plt.colorbar(contour)
# Set the colorbar label
cbar.set_label('Sea Surface Temperature (SST) in °C')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('NOAA ERSST V5 - January Climatology')


plt.savefig('sst_tutorial.png', bbox_inches='tight', dpi=800)

plt.show()
```
